```python

def currency (data):
    currency_symbol =['$', '€','¥','£','₹']
    for i in currency_symbol:
        if i in data:
            return 1
    return 0  
```

```python
#Data cleaning process
import nltk
import re 
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('stopwords')
nltk.download('wordnet')

corpus = []
wnl = WordNetLemmatizer()
for sms in spam_data ['message']:  # Assuming 'smsdata' is your DataFrame with an 'SMS' column
    message = re.sub(pattern='[^a-zA-Z]', repl=' ', string=sms)  # Replacing non-alphabetic characters with a space
    message = message.lower()
    words = message.split()
    filtered_words = [word for word in words if word not in set(stopwords.words('english'))]
    lemmatized_words = [wnl.lemmatize(word) for word in filtered_words]
    message = ' '.join(lemmatized_words)
    corpus.append(message)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features= 500)
vectors = tfidf.fit_transform(corpus).toarray()
feature_names = tfidf.get_feature_names_out()

```

```python
x = pd.DataFrame(vectors , columns= feature_names)
y = spam_data['class']
```

```python
x
```

```python
y
```

```python

# Create TF-IDF features
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import classification_report,confusion_matrix
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_test
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
cv = cross_val_score(mnb, x,y , scoring='f1',cv=10)
print(round(cv.mean(),3))
print(round(cv.std(),3))
```

```python
mnb.fit(X_train,Y_train)
y_pred = mnb.predict(X_test)
```

```python
y_pred
```

```python
Y_test
```

```python

print(classification_report(Y_test,y_pred))
```

```python
cm = confusion_matrix(Y_test,y_pred)
cm
```

```python

def predict_spam(sms, tfidf_vectorizer,model):
    message = re.sub(pattern='[^a-zA-Z]', repl=' ', string=sms)
    message = message.lower()
    words = message.split()
    filtered_words = [word for word in words if word not in set(stopwords.words('english'))]
    lemmatized_words = [wnl.lemmatize(word) for word in filtered_words]  # Lemmatize each word
    message= ' '.join(lemmatized_words)
    temp = tfidf_vectorizer.transform([message]).toarray()
    prediction = model.predict(temp) 


    return prediction[0]

```
